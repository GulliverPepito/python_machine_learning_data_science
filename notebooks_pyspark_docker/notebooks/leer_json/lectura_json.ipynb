### Módulo pyspark sql

```python
from pyspark import SparkContext
from pyspark.sql import SQLContext

from pyspark.sql.types import *
from pyspark.sql import functions

spark= SparkSession.builder.appName("Leer un fichero json con Spark SQL").enableHiveSupport().getOrCreate()

sc = spark.sparkContext
sqlCtx = SQLContext(sc)

version =spark.version
print('Version',sqlCtx)

data = spark.read.json("estudiantes.json")
data
```

```python
data.count()
```

```python
data.collect()
```

```python
data.show()
```

```python
data.printSchema()
```

```python
(data.filter(data['nombre'] != 'null')
   .filter(data['nota'] > 8)
   .select(['nombre', 'apellidos', 'nota'])
   .show())
```

```python
data.registerTempTable("estudiantes")
```

```python
sqlContext.sql("""
    SELECT nombre,apellidos, nota 
    FROM estudiantes 
    WHERE nota >8""").show()
```

```python
sqlContext.table("estudiantes").collect()
```

```python
row = sqlContext.sql("SELECT * FROM estudiantes").first()
row
```
